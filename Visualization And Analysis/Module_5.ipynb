```python
#Remove duplicates
df = df.drop_duplicates()
df
```

```python
# Average price (MSRP) by car brand
avg_price = df.groupby('Make')['MSRP'].mean()
print(avg_price)
```

```python
avg_price.plot(kind='bar', color='red', figsize=(8,5))
plt.title("Average Car Price by Brand", fontsize=14)
plt.xlabel("Car Brand", fontsize=12)
plt.ylabel("Average Price (MSRP)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#Cars costing more than 25000
expensive_cars = df[df['MSRP'] > 25000]
expensive_cars
```

```python
brand_info = pd.DataFrame({
    'Make': ['Toyota','Honda','BMW'],
    'Country': ['Japan','Japan','Germany']
})

merged_df = pd.merge(df, brand_info, on='Make')
merged_df
```

```python
sns.lineplot(x='Year', y='MSRP', data=df, marker='o')
plt.title("Car Price Trend by Year")
plt.show()
```

```python
sns.histplot(df['MSRP'], bins=10, kde=True, color='orange')
plt.title("Car Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Engine HP'], df['MSRP'], color='orange', edgecolor='black')
plt.title('Engine Horsepower vs Price')
plt.xlabel('Engine HP')
plt.ylabel('MSRP ($)')
plt.grid(True)
plt.show()

```

```python
summary = df.groupby('Make').agg({'MSRP': ['count', 'mean', 'max', 'min']})
print(summary)
```

```python
pivot = pd.pivot_table(df, values='MSRP', index='Year', columns='Make', aggfunc='mean')
pivot
```
